```python
import cv2
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('/home/shreyanka/Documents/Clique/haarcascade_frontalface_default.xml')
```

```python
# It will be using webcamera
cap = cv2.VideoCapture(0)
while True:
    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,1.3, 5)
    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y) , (x+w,y+h), (255,0,0), 2)
        
    cv2.imshow('img', img)
    k = cv2.waitKey(30) & 0xff
    if k==5:
        break
cap.release()
cv2.destroyAllWindows()
```
